```python
def compute(obj):
    obj=ast.literal_eval(obj)
    new_overview=[]
    for dict_ele in obj:
        new_overview.append(dict_ele["name"])
    return new_overview

def getTop(obj):
    obj=ast.literal_eval(obj)
    top_casts=[]
    mini=min(len(obj),3)
    for i in range(0,mini):
         top_casts.append(obj[i]['name'])
    return top_casts

    
def callingFunction(columns):
    for col in columns:
        df[col]=df[col].apply(compute)
```

```python
def clean_up_sentence(sentence): # conerts all to loercase
    sentence_words = nltk.word_tokenize(sentence)
    sentence_words = [stemmer.stem(word.lower()) for word in sentence_words]
    return sentence_words
```

```python
def extract_keywords(sentence, num_keywords=5): # removes stopwords and then->tokenisation->stemming
    words = word_tokenize(sentence.lower())
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    stemmer = PorterStemmer()
    words = [stemmer.stem(word) for word in words]
    return words
```

### Vectorise 5k unique keywords

```python
from sklearn.feature_extraction.text import CountVectorizer
CV=CountVectorizer(max_features=5000)
vectors=CV.fit_transform(df2['tags']).toarray()
from sklearn.metrics.pairwise import cosine_similarity
obj=cosine_similarity(vectors)  
```
